```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol
ticker_symbol = 'ALC'

# Fetch the ticker data
ticker = yf.Ticker(ticker_symbol)

# Get annual financial statements
income_statement_annual = ticker.financials if ticker.financials is not None else pd.DataFrame()
balance_sheet_annual = ticker.balance_sheet if ticker.balance_sheet is not None else pd.DataFrame()
cash_flow_annual = ticker.cashflow if ticker.cashflow is not None else pd.DataFrame()

# Get quarterly financial statements
income_statement_quarterly = ticker.quarterly_financials if ticker.quarterly_financials is not None else pd.DataFrame()
balance_sheet_quarterly = ticker.quarterly_balance_sheet if ticker.quarterly_balance_sheet is not None else pd.DataFrame()
cash_flow_quarterly = ticker.quarterly_cashflow if ticker.quarterly_cashflow is not None else pd.DataFrame()

# Function to extract financial metrics from statements


def extract_metrics(statement):
    metrics = {}
    try:
        metrics['Total Revenue'] = statement.loc['Total Revenue'] if not statement.empty else None
    except KeyError:
        metrics['Total Revenue'] = None

    try:
        metrics['Gross Profit'] = statement.loc['Gross Profit'] if not statement.empty else None
    except KeyError:
        metrics['Gross Profit'] = None

    try:
        metrics['Operating Income'] = statement.loc['Operating Income'] if not statement.empty else None
    except KeyError:
        metrics['Operating Income'] = None

    try:
        metrics['Net Income'] = statement.loc['Net Income'] if not statement.empty else None
    except KeyError:
        metrics['Net Income'] = None

    return metrics


# Extract income statement metrics
income_statement_metrics_annual = extract_metrics(income_statement_annual)
income_statement_metrics_quarterly = extract_metrics(
    income_statement_quarterly)

# Extract balance sheet metrics (you can add more metrics as needed)


def extract_balance_sheet_metrics(statement):
    metrics = {}
    try:
        metrics['Total Assets'] = statement.loc['Total Assets'] if not statement.empty else None
    except KeyError:
        metrics['Total Assets'] = None

    try:
        metrics['Total Liabilities'] = statement.loc['Total Liabilities Net Minority Interest'] if not statement.empty else None
    except KeyError:
        metrics['Total Liabilities'] = None

    try:
        metrics['Cash and Cash Equivalents'] = statement.loc['Cash'] if not statement.empty else None
    except KeyError:
        metrics['Cash and Cash Equivalents'] = None

    return metrics


# Extract balance sheet metrics
balance_sheet_metrics_annual = extract_balance_sheet_metrics(
    balance_sheet_annual)
balance_sheet_metrics_quarterly = extract_balance_sheet_metrics(
    balance_sheet_quarterly)

# Extract cash flow statement metrics


def extract_cash_flow_metrics(statement):
    metrics = {}
    try:
        metrics['Operating Cash Flow'] = statement.loc['Operating Cash Flow'] if not statement.empty else None
    except KeyError:
        metrics['Operating Cash Flow'] = None

    try:
        metrics['Capital Expenditures'] = statement.loc['Capital Expenditure'] if not statement.empty else None
    except KeyError:
        metrics['Capital Expenditures'] = None

    return metrics


# Extract cash flow metrics
cash_flow_metrics_annual = extract_cash_flow_metrics(cash_flow_annual)
cash_flow_metrics_quarterly = extract_cash_flow_metrics(cash_flow_quarterly)

# Combine all the metrics into DataFrames for annual and quarterly data
metrics_annual = {
    **income_statement_metrics_annual,
    **balance_sheet_metrics_annual,
    **cash_flow_metrics_annual
}

metrics_quarterly = {
    **income_statement_metrics_quarterly,
    **balance_sheet_metrics_quarterly,
    **cash_flow_metrics_quarterly
}

# Create DataFrames for annual and quarterly metrics
metrics_annual_df = pd.DataFrame(metrics_annual)
metrics_quarterly_df = pd.DataFrame(metrics_quarterly)

# Transpose the DataFrames for better readability
alc_metrics_annual_df = metrics_annual_df.T
alc_metrics_quarterly_df = metrics_quarterly_df.T

# Display the metrics
print("Selected Annual Financial Metrics for", ticker_symbol)
print(alc_metrics_annual_df)

print("\nSelected Quarterly Financial Metrics for", ticker_symbol)
print(alc_metrics_quarterly_df)
```

```python
import pandas as pd
import openai
import os

# Step 1: Read the Excel file from the given file path
file_path = r'C:\Users\camjo\OneDrive\Desktop\Capstone Project\financial_metrics.xlsx'
metrics_df = pd.read_excel(file_path)

# Step 2: Convert the DataFrame to CSV format (excluding the index)
metrics_csv = metrics_df.to_csv(index=False)

# Step 3: Define the OpenAI API key securely (use environment variables)
openai.api_key = ""

# Ensure the API key is set
if openai.api_key is None:
    raise ValueError(
        "OpenAI API key not found. Please set the OPENAI_API_KEY environment variable.")

# Step 4: Construct the detailed prompt
detailed_prompt = f"""
I am a financial analyst at Alcon (ALC), and I am currently conducting a detailed financial analysis to gather insights for an internal report.
I am focusing on key financial metrics for Alcon as well as its competitors Bausch + Lomb Corporation (BLCO), RxSight, Inc. (RXST), and CooperVision (COO).
Specifically, I am analyzing the financial statements, including income statements, balance sheets, and cash flow statements for these companies over the two most recent fiscal years.

For this analysis, I have extracted key metrics such as total revenue, gross profit, operating income, net income, EBITDA, total assets, total liabilities, total debt, operating cash flow, free cash flow, and capital expenditures.
These metrics will help me compare the financial performance, liquidity, profitability, and capital efficiency of Alcon relative to its competitors.

The goal of this analysis is to identify strengths, weaknesses, and areas of opportunity for Alcon and determine how the company is performing in relation to industry peers.
Additionally, I aim to evaluate the companies' cash flow management, capital expenditures, and debt levels to provide a comprehensive outlook on their financial health and sustainability.

Here is the financial data in CSV format for the last two fiscal years for ALC, BLCO, and RXST:

{metrics_csv}

Please help me interpret the data and provide actionable insights based on trends, comparisons, and potential strategic recommendations for Alcon.
"""

# Step 5: Send the prompt to OpenAI's ChatCompletion API
response = openai.ChatCompletion.create(
    model="gpt-4",
    messages=[
        {"role": "system", "content": "You are an assistant that helps analyze financial data."},
        {"role": "user", "content": detailed_prompt}
    ]
)

# Step 6: Print the response from ChatGPT
print(response['choices'][0]['message']['content'])
```
